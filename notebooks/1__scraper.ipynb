# House

### 1976 - House

```python
url = requests.get('https://en.wikipedia.org/wiki/United_States_House_of_Representatives_elections,_1976')

soup = BeautifulSoup(url.text, 'lxml')
tables = soup.findAll('table', attrs = {'class' : 'wikitable'})
tables.pop(0);
tables.pop(0);
```

```python
def scrapey(table):
    race_list = []
    rows = table.findAll('tr')
    rows.pop(0);
    for row in rows:
        new_race = {}
        cells = row.findAll('td')
        print(cells[0].text)
        if len(cells) < 6:
            print('problem with', cells[0].text)
            continue
        new_race['district'] = cells[0].text
        new_race['inc'] = cells[1].text
        new_race['inc_party'] = cells[2].text
        candidates = cells[5].text.split('%')
        new_race['candidates'] = candidates
        new_race['winner'] = cells[5].text.split('%')[0]
        if len(candidates) == 1:
            new_race['loser'] = np.nan
        else:
            new_race['loser'] = cells[5].text.split('%')[1]
        race_list.append(new_race)
    return pd.DataFrame(race_list)
```

```python
house_races_76 = scrapey(tables[0])
for table in tables[1:]:
    house_races_76 = house_races_76.append(scrapey(table))
print(house_races_76.shape)
house_races_76['year'] = 1976
house_races_76.head()
```

```python
house_races_76.to_csv('./house_races_76.csv')
```

### 1976 Senate

```python
def sen_scrape(table):
    race_list = []
    rows = table.findAll('tr')
    rows.pop(0);
    rows.pop(0);
    for row in rows:
        new_race = {}
        cells = row.findAll('td')
#         print(cells[0].text)
        new_race['state'] = cells[0].text
        new_race['inc'] = cells[1].text
        new_race['inc_party'] = cells[2].text
        candidates = cells[5].text.split('%')
        new_race['candidates'] = candidates
        new_race['winner'] = candidates[0]
        
        
        race_list.append(new_race)
    return pd.DataFrame(race_list)
```

```python
url = requests.get('https://en.wikipedia.org/wiki/United_States_Senate_elections,_1976')
soup = BeautifulSoup(url.text, 'lxml')
tables = soup.findAll('table', attrs = {'class' : 'wikitable'})
# tables.pop(0);
# tables.pop(0);
# tables.pop(0);
# tables.pop(0);
table = tables[4]
senate_1976 = sen_scrape(table)
senate_1976['year'] = 1976
print(senate_1976.shape)
senate_1976.head()
```

```python
def sen_scrape2(table):
    race_list = []
    rows = table.findAll('tr')
    rows.pop(0);
    rows.pop(0);
    for row in rows:
        new_race = {}
        cells = row.findAll('td')
#         print(cells[0].text)
        new_race['state'] = row.find('th').text
        new_race['inc'] = cells[0].text
        new_race['inc_party'] = cells[1].text
        candidates = cells[4].text.split('%')
        new_race['candidates'] = candidates
        new_race['winner'] = candidates[0]
        
        
        race_list.append(new_race)
    return pd.DataFrame(race_list)
```

```python
def gov_scrape(table):
    race_list = []
    rows = table.findAll('tr')
    rows.pop(0);
#     rows.pop(0);
    for row in rows:
        new_race = {}
        cells = row.findAll('td')
#         print(cells[0].text)
        new_race['state'] = cells[0].text
        new_race['inc'] = cells[1].text
        new_race['inc_party'] = cells[2].text
        new_race['inc_results'] = cells[3].text
        candidates = cells[4].text.split('%')
        new_race['candidates'] = candidates
        new_race['winner'] = candidates[0]
        
        
        race_list.append(new_race)
    return pd.DataFrame(race_list)
```

```python
base_url = 'https://en.wikipedia.org/wiki/United_States_gubernatorial_elections,_'
years1 = []
years3 = []
for i in range(1977, 2011):
    years1.append(str(i))

years2 = ['2011']
    
for i in range(2012, 2016):
    years3.append(str(i))
    
years4 = ['2016']
years5 = ['2017']
```

```python
for year in years1:
    print("Scanning for..", year)
    url = requests.get(base_url + year)
    soup = BeautifulSoup(url.text, 'lxml')
    tables = soup.findAll('table', attrs = {'class' : 'wikitable'})
    table = tables[0]
    new_gov_year = gov_scrape(table)
    new_gov_year['year'] = year
    governors = governors.append(new_gov_year)
```
